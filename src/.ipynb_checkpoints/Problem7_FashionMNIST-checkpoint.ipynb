## DBSCAN on Real Data

```python
# Importing the required libraries

import struct
from pathlib import Path

import numpy as np
import pandas as pd
from tqdm import tqdm

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
class MinMaxScalar:

	def __init__(self):
		self.has_run_fit = False

		self.min = None
		self.max = None


	def fit(self, X: np.array) -> None:
		self.min = np.min(X)
		self.max = np.max(X)

		self.has_run_fit = True


	def transform(self, X: np.array) -> np.array:
		if not self.has_run_fit:
			raise Exception("Have to call the .fit() method before calling the .transform() method")
		
		X = ((X - self.min) / self.max)
		return X


	def fit_transform(self, X: np.array) -> np.array:
		self.fit(X)
		X = self.transform(X)

		return X
```

```python
class FashionMNISTDataParser:

    # File URLS
    BASE_DIR = Path.cwd().parent
    MNIST_DATA_DIR = BASE_DIR / "data" / "fashion-mnist"

    TRAIN_DATA_URL = MNIST_DATA_DIR / "train-images-idx3-ubyte"
    TRAIN_LABELS_URL = MNIST_DATA_DIR / "train-labels-idx1-ubyte"

    TEST_DATA_URL = MNIST_DATA_DIR / "t10k-images-idx3-ubyte"
    TEST_LABELS_URL = MNIST_DATA_DIR / "t10k-labels-idx1-ubyte"

    IDX_TO_CLASS_LABELS = {
        0: "T-Shirt/Top",
        1: "Trouser",
        2: "Pullover",
        3: "Dress",
        4: "Coat",
        5: "Sandal",
        6: "Shirt",
        7: "Sneaker",
        8: "Bag",
        9: "Ankle Boot"
    }


    def __init__(self):
        self.scalar = MinMaxScalar()


    def load_dataset(self, dataset_url: str) -> np.array:       
        with open(dataset_url,'rb') as data_file:
            magic, size = struct.unpack(">II", data_file.read(8))
            nrows, ncols = struct.unpack(">II", data_file.read(8))
            data = np.frombuffer(data_file.read(), dtype=np.dtype(np.uint8).newbyteorder('>'))
            data = data.reshape((size, -1))
        
        return data


    def load_labels(self, labels_url: str) -> np.array:
        with open(labels_url,'rb') as labels_file:
            magic, size = struct.unpack('>II', labels_file.read(8))
            label = np.frombuffer(labels_file.read(), dtype=np.dtype(np.uint8).newbyteorder('>'))
            return label


    @staticmethod
    def show_image(image: np.array, title: str) -> None:
        plt.figure(figsize=(12, 8))
        plt.imshow(image, cmap=plt.get_cmap("gray"))
        plt.title(title)
        plt.show()

    
    def parse_data(self, verbose: int = 1, sample_data: bool = False) -> tuple[np.array, np.array, np.array, np.array]:
        X_train = self.load_dataset(self.TRAIN_DATA_URL)
        y_train = self.load_labels(self.TRAIN_LABELS_URL)

        X_test = self.load_dataset(self.TEST_DATA_URL)
        y_test = self.load_labels(self.TEST_LABELS_URL)

        if sample_data:
            n_train_samples = 10000
            train_rand_idx = np.random.choice(X_train.shape[0], n_train_samples, replace=False)
            X_train = X_train[train_rand_idx]
            y_train = y_train[train_rand_idx]

            n_test_samples = 5000
            test_rand_idx = np.random.choice(X_test.shape[0], n_test_samples, replace=False)
            X_test = X_test[test_rand_idx]
            y_test = y_test[test_rand_idx]
		
        self.scalar.fit(X_train)
		
        X_train_scaled = self.scalar.transform(X_train)
        X_test_scaled = self.scalar.transform(X_test)

        if verbose:
            rand_train_idx = np.random.randint(low=0, high=X_train.shape[0])
            title = f"MNIST Training Image - Label: {self.IDX_TO_CLASS_LABELS[y_train[rand_train_idx]]}"
            self.show_image(X_train_scaled[rand_train_idx].reshape(28, 28), title)

            rand_test_idx = np.random.randint(low=0, high=X_test.shape[0])
            title = f"MNIST Testing Image - Label: {self.IDX_TO_CLASS_LABELS[y_test[rand_test_idx]]}"
            self.show_image(X_test_scaled[rand_test_idx].reshape(28, 28), title)

        return (X_train_scaled, y_train, X_test_scaled, y_test)
```

```python
data_parser = FashionMNISTDataParser()
X_train_scaled, y_train, X_test_scaled, y_test = data_parser.parse_data(verbose=0, sample_data=False)

print(f"Shape of Training Data = {X_train_scaled.shape}")
print(f"Shape of Training Labels = {y_train.shape}")

print(f"Shape of Testing Data = {X_test_scaled.shape}")
print(f"Shape of Testing Labels = {y_test.shape}")
```

```python
from sklearn.model_selection import train_test_split

_, X_train_scaled, _, y_train = train_test_split(X_train_scaled, y_train, test_size=0.2, random_state=0, stratify=y_train)
```

```python
from scipy.spatial.distance import cdist


def euclidean_distance(points) -> np.array:
    distances = np.empty(shape=(points.shape[0], points.shape[0]), dtype=np.float16)

    chunk_size = 1000
    idx = 0

    for idx in tqdm(range(0, points.shape[0], chunk_size), desc="Calculating distance"):
        points_subset = points[idx : idx + chunk_size, :]
        
        distances[idx : idx + chunk_size, :] = cdist(points_subset, points)
        idx += chunk_size

    return distances

def create_neighbors(X: np.array, eps: float) -> pd.DataFrame:
    distance_matrix = euclidean_distance(X)

    # print(f"distance_matrix max = {distance_matrix.max()}, distance_matrix min = {distance_matrix.min()}")

    df = pd.DataFrame()
    for idx in tqdm(range(distance_matrix.shape[0]), desc="Creating neighbors"):
        neighbors_array = np.where(distance_matrix[idx, :] <= eps)[0]
        
        df.loc[idx, "neighbors"] = ",".join(
            list(map(str, neighbors_array))
        )

        df.loc[idx, "num_neighbors"] = neighbors_array.shape[0]

        df.loc[idx, "pt"] = idx

    df = df.astype({"num_neighbors": int, "pt": int})

    df["cluster"] = np.nan
    
    return df

df = create_neighbors(X_train_scaled, eps=7.5)
```

```python
class CustomDBSCAN:

    def __init__(self, eps: float, min_pts: int):
        self.eps = eps
        self.min_pts = min_pts

    
    def fit_predict(self, df: pd.DataFrame):        
        cluster_id = 0
        
        has_set_cluster = False
        num_points_without_cluster = df["cluster"].isna().sum()
        
        while num_points_without_cluster > 0:
            random_starting_point = df[df["cluster"].isna()].sample(n=1)

            queue = []
            queue.append(int(random_starting_point["pt"]))

            pts_visited = set()
            pts_visited.add(int(random_starting_point["pt"]))

            has_set_cluster = False

            while len(queue) > 0:
                point = queue.pop(0)

                # Setting the cluster id to this point
                if has_set_cluster or df[df["pt"] == point]["num_neighbors"].item() >= self.min_pts:
                    df.loc[df["pt"] == point, "cluster"] = cluster_id
            
                    # Adding all the neighbors to the queue
                    for neighbor in df[df["pt"] == point]["neighbors"].item().split(","):
                        if int(neighbor) not in pts_visited:
                            queue.append(int(neighbor))
                            pts_visited.add(int(neighbor))

                    has_set_cluster = True

                else:
                    df.loc[df["pt"] == point, "cluster"] = -1

            if has_set_cluster:
                cluster_id += 1

            num_points_without_cluster = df["cluster"].isna().sum()

        df["cluster"] = df["cluster"].astype(int)
        
        return df
```

```python
dbscan = CustomDBSCAN(eps=7.5, min_pts=5)
```

```python
result_df = dbscan.fit_predict(df.copy())
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

standardized_data = StandardScaler().fit_transform(X_train_scaled)

pca = PCA(n_components=2)

pca_data = pca.fit_transform(standardized_data)
```

```python
percentage_var_explained = pca.explained_variance_ / np.sum(pca.explained_variance_);
cum_var_explained = np.cumsum(percentage_var_explained)
```
